```python
#    Six Supersteps
# 1. Retrieve: Load and structure raw data from the source(like CSV, SQL, Excel, XML files).
# 2. Assess: Clean and enhance the data for quality(by checking for missing values, duplicates, and data consistency).
# 3. Process: Organize data into a structured format.
# 4. Transform: Aggregate and structure data for analysis.
# 5. Organize: Create specialized subsets of data for specific uses.
# 6. Report: Generate and visualize reports to present insights.
```

```python
# Include all necesssary imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Modules imported successful!")
```

```python
# 1. Retrieve 
data = pd.read_csv('Bengaluru_Restaurants.csv') # Load the dataset

# Display some basic info about the dataset
data.head(3) # Display the first few rows 
```

```python
# Display some basic info about the dataset
data_info = data.info() 
data_info
```

```python
# 2. Assess
# Remove unnecessary columns that may not contribute much to analysis

# Columns like 'description' may not be useful as they have high missing values and it's free text data
# Removing 'address', 'localAddress', etc. that are unlikely to impact analysis significantly
# Removing 'addressObj/country', 'addressObj/state' as we already know it begaluru databset and benfulaure is in India
cleaned_data = data.drop(columns=['address', 'addressObj/country', 'addressObj/state', 'description', 'localAddress', 'phone'])

# Renaming column names
cleaned_data.rename({'addressObj/postalcode': 'pincode','DietaryRestrictions': 'dietaryRestrictions', 'Dishes': 'dishes', 'Features': 'features','Meal Type': 'mealType'}, axis=1, inplace=True)

cleaned_data.head()
```

```python
# Check for missing values in the dataset
missing_values = cleaned_data.isnull().sum()
missing_values
```

```python
# 3. Process
# Handle missing values:
# For 'latitude' and 'longitude', we'll drop rows where these are missing (since geolocation is important)
# For 'rankingDenominator', 'rankingPosition', and 'rating', handle based on a strategy like mean/mode imputation or remove rows with too many missing values.
cleaned_data = cleaned_data.dropna(subset=['latitude', 'longitude'])

# Fill missing values in 'rating', 'rankingPosition', 'rankingDenominator' and 'rawRanking' with the mean of their respective columns
cleaned_data['rating'] = cleaned_data['rating'].fillna(cleaned_data['rating'].mean())
cleaned_data['rankingPosition'] = cleaned_data['rankingPosition'].fillna(cleaned_data['rankingPosition'].mean())
cleaned_data['rankingDenominator'] = cleaned_data['rankingDenominator'].fillna(cleaned_data['rankingDenominator'].mean())
cleaned_data['rawRanking'] = cleaned_data['rawRanking'].fillna(cleaned_data['rawRanking'].mean())

# Check the remaining missing values after cleaning
cleaned_missing_values = cleaned_data.isnull().sum()
cleaned_missing_values
```

```python
# 4. Transform
# Define a function to categorize cuisines into broader groups
def categorize_cuisine(cuisine_str):
    cuisine_str = cuisine_str.lower()

    categories = {
        'indian': 'Indian',
        'chinese': 'Chinese',
        'asian': 'Asian',
        'italian': 'Italian',
        'barbecue': 'Barbecue',
        'american': 'American',
        'mediterranean': 'Mediterranean',
        'japanese': 'Japanese',
        'mexican': 'Mexican',
        'european': 'European',
        'french': 'French',
        'thai': 'Thai',
        'pub': 'Pub',
        'street food': 'Street Food',
        'fast food': 'Fast Food',
        'fusion': 'Fusion',
        'pizza': 'Pizza',
        'steakhouse': 'Steakhouse',
        'sushi': 'Sushi',
        'seafood': 'Seafood',
        'grill': 'Grill',
        'bar': 'Bar',
        'cafe': 'Cafe',
        'gastropub': 'Gastropub',
        'beer restaurants': 'Beer Restaurants',
        'wine bar': 'Wine Bar',
        'dining bars': 'Dining Bars',
        'contemporary': 'Contemporary',
        'healthy': 'Healthy',
        'international': 'International',
        'south american': 'South American',
        'central european': 'Central European',
        'northwestern chinese': 'Northwestern Chinese',
        'central asian': 'Central Asian',
        'afghan': 'Afghan',
        'arabic': 'Arabic',
        'burmese': 'Burmese',
        'cambodian': 'Cambodian',
        'cantonese': 'Cantonese',
        'korean': 'Korean',
        'malaysian': 'Malaysian',
        'nepali': 'Nepali',
        'persian': 'Persian',
        'portuguese': 'Portuguese',
        'szechuan': 'Szechuan',
        'shanghai': 'Shanghai',
        'singaporean': 'Singaporean',
        'sri lankan': 'Sri Lankan',
        'taiwanese': 'Taiwanese',
        'vietnamese': 'Vietnamese',
        'wine bar': 'Wine Bar'
    }

    for keyword, category in categories.items():
        if keyword in cuisine_str:
            return category

    return 'Generic'  # Default category for unrecognized cuisines

# Apply the function to the 'cuisine' column
cleaned_data['cuisine_category'] = cleaned_data['cuisine'].apply(categorize_cuisine)

# Show the first few rows and last few with the new 'cuisine_category' column
cleaned_data[['cuisine', 'cuisine_category']].head(), cleaned_data[['cuisine', 'cuisine_category']].tail()
```

```python
# Displaying some basic stats
cleaned_data.describe()
```

```python
# 5. Organize
cleaned_data.to_csv('Bengaluru_Restaurants_cleaned_data.csv')
```

```python
# 6. Report
# To visualize the distribution of restaurant ratings
plt.figure(figsize=(8, 6))
sns.histplot(cleaned_data['rating'], bins=20, kde=True) # kde adds a Kernel Density Estimation line
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.legend(labels=['Kernel Density Estimate line', 'Ratings Distribution'])  
plt.show()
```

```python
# You can explore which cuisines are the most popular

top_cuisines = cleaned_data['cuisine_category'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values)
plt.title('Top 10 Cuisines')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming cleaned_data is a pandas DataFrame
stats = cleaned_data.describe()

# Plot using Matplotlib
plt.figure(figsize=(12, 8))
plt.bar(stats.columns, stats.loc['mean'], label='Mean', alpha=0.7)
plt.errorbar(stats.columns, stats.loc['mean'], yerr=stats.loc['std'], fmt='o', color='red', label='Standard Deviation')

plt.title('Basic Statistics: Mean and Standard Deviation')
plt.xlabel('Features')
plt.ylabel('Values')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Plot using Seaborn (Heatmap for overall summary)
plt.figure(figsize=(12, 8))
sns.heatmap(stats.T, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Statistical Summary (describe())')
plt.show()
```
