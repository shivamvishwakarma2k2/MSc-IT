```python
import matplotlib.pyplot as plt

fig, graph = plt.subplots()

fruits = ['grapes','blueberry','apple','blue berry', 'Orange']
counts = [ 49, 25, 55, 14, 44]

fruit_colors = ['green', 'purple', 'red', 'blue', 'orange']

graph.bar(fruits, counts, label = fruit_colors, color = fruit_colors)
graph.set_xlabel('Fruits')
graph.set_ylabel('Supply (in KGs)')
graph.set_title('Fruit supply in Quantity')
graph.legend(title='Color to Fruits')
plt.show()
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15,30,45,10]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

```python
import matplotlib.pyplot as plt

labels = 'IT', 'HR', 'Sales', 'Marketing'
sizes = [20, 10, 40, 30]  # Percentage of employees in each department
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']  # Custom colors
explode = (0, 0, 0.1, 0)  # "explode" the 3rd slice (Sales)

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)

ax.axis('equal')  
plt.show()
```

```python
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)
fig, ax = plt.subplots()
ax.plot(t, s)
ax.set(xlabel='time (s)', ylabel='voltage (mV)', title='About as simple as it gets, folks')
ax.grid()
fig.savefig("D:\\MSc-IT\\Part 1\\SEM 1\\IDS\\Practicals\\Practical 8\\pie_img.png")
plt.show()
```
